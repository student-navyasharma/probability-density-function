```python



import numpy as np
import pandas as pd
from scipy.optimize import minimize
from google.colab import files


uploaded = files.upload()


file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name, encoding='latin1')



print("Columns in dataset:", df.columns)


# Keep only no2 column (lowercase)
x = pd.to_numeric(df["no2"], errors='coerce').dropna().values





ROLL_NUMBER = 102303313

a_r = 0.05 * (ROLL_NUMBER % 7)
b_r = 0.3 * ((ROLL_NUMBER % 5) + 1)

print("a_r:", a_r)
print("b_r:", b_r)

# Transformation
z = x + a_r * np.sin(b_r * x)





def neg_log_likelihood(params, data):
    lam, mu, c = params


    if lam <= 0 or c <= 0:
        return 1e10

    log_likelihood = np.log(c) - lam * (data - mu)**2
    return -np.sum(log_likelihood)


init_params = [0.01, np.mean(z), 0.01]


bounds = [(1e-6, None), (None, None), (1e-6, None)]

result = minimize(
    neg_log_likelihood,
    init_params,
    args=(z,),
    method='L-BFGS-B',
    bounds=bounds
)

lam_hat, mu_hat, c_hat = result.x




print("\nEstimated Parameters:")
print("Lambda (λ):", lam_hat)
print("Mu (μ):", mu_hat)
print("c:", c_hat)
```
